```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# File to Load
game_path = "purchase_data.csv"

# Read the csv and store into Pandas DataFrame
game_df = pd.read_csv(game_path)
game_df.head()
```

```python
#Display the total number of players
total_players = len(game_df["SN"].unique())
players_df= pd.DataFrame({"Total Players":[total_players]})
players_df
```

```python
#Calculate the number of unique items,average price, total number of purchase, and total revenue
number_of_unique_items = len(game_df["Item ID"].unique())
average_price = game_df["Price"].mean()
tot_number_of_purchases = game_df["Purchase ID"].count()
total_revenue = game_df["Price"].sum()

#Create a summary data frame to hold the results
summary_df = pd.DataFrame({"Number of Unique Items":[number_of_unique_items],
                           "Average Price":[average_price], 
                           "Number of Purchases": [tot_number_of_purchases], 
                           "Total Revenue":[total_revenue]})
#Give the displayed data cleaner formatting
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

summary_df
```

```python
#group by gender and find total count and percentage 
Gender = game_df.groupby("Gender")
count_SN = Gender.nunique()["SN"]
players = round(count_SN / total_players * 100, 2)

# Create a summary data frame to hold the results
gender_df = pd.DataFrame({"Total Count": count_SN,
                         "Percentage of Players":players})

#Give the displayed data cleaner formatting
gender_df=gender_df.sort_values("Total Count", ascending=False)


gender_df
```
